### 先从玉米开始

```python
import someFunction as sF
import pandas as pd
```

```python
oil,d5,dp5,dp6 = sF.getOdata()
test_data,test_lable,train_data,train_lable = sF.getTestData(oil,d5,dp5,dp6 )
train_data = train_data.reshape(60, 700)
test_data = test_data.reshape(20, 700)
```

```python
for i in range(200):
    a = []
    pls = PLSRegression(n_components=i+1,scale=False)
    pls.fit(train_data,train_lable)
    res_train = pls.predict(train_data)
    res_test = pls.predict(test_data)
    rmsec = sF.calculate_RMSE(res_train,train_lable) ## 训练集上的RMSE
    rmsep = sF.calculate_RMSE(res_test,test_lable)  ## 测试集上的RMSE
    r_2_t = sF.calculate_R21(res_train,train_lable)## 训练集上的R_2
    r_2_p = sF.calculate_R21(res_test,test_lable)## 测试集上得R_2
    a.append(i)
    a.append(rmsec)
    a.append(r_2_t)
    a.append(rmsep)
    a.append(r_2_p)
    print(a)
    data = pd.DataFrame([a])
    data.to_csv('pls-yumi.csv',mode='a',index=False,header=False)
```
