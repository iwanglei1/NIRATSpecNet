### BP玉米

```python
import someFunction as sF
import pandas as pd
from sklearn.neural_network import MLPRegressor
oil,d5,dp5,dp6 = sF.getOdata()
test_data,test_lable,train_data,train_lable = sF.getTestData(oil,d5,dp5,dp6 )
train_data = train_data.reshape(60, 700)
test_data = test_data.reshape(20, 700)
for i in range(1):
    a = []
    mlp = MLPRegressor(random_state=2, max_iter=5000,hidden_layer_sizes=(650,600,550,500,450,400,350,300,250,200,150,100,50))##650,600,550,500,450,400,350,300,250,200,150,100,50
    mlp.fit(train_data, train_lable)
    res_train = mlp.predict(train_data)
    res_test = mlp.predict(test_data)
    rmsec = sF.calculate_RMSE(res_train,train_lable) ## 训练集上的RMSE
    rmsep = sF.calculate_RMSE(res_test,test_lable)  ## 测试集上的RMSE
    r_2_t = sF.calculate_R21(res_train,train_lable)## 训练集上的R_2
    r_2_p = sF.calculate_R21(res_test,test_lable)## 测试集上得R_2
    a.append(i)
    a.append(rmsec)
    a.append(r_2_t)
    a.append(rmsep)
    a.append(r_2_p)
    data = pd.DataFrame([a])
    data.to_csv('bp-yumi.csv',mode='a',index=False,header=False)
```
